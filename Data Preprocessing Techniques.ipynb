```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

np.random.seed(42)
def createdata():
    data = {
        'Age':np.random.randint(18,70,size=20),
        'Salary':np.random.randint(30000,120000,size=20),
        'Purchased': np.random.choice([0,1],size=20),
        'Gender':np.random.choice(['Male','Female'],size=20),
        'City':np.random.choice(['New York','San Francisco','Los Angeles'],size=20)
        }
    df = pd.DataFrame(data)
    return df
    
df=createdata()
df.head(10)
    
```

When you have categorical variables in data

Label, ordinal, and one-hot encoding are all techniques for representing categorical data in a machine learning model:

    Label encoding
    Converts labels or words into numeric values. This technique is best for categorical features with only two categories. Label encoding is simple, but it can introduce false ordinal relationships.
    
    One-hot encoding
    Represents categorical variables as binary vectors. This technique is best for nominal categorical variables, where the categories have no inherent order or relationship. One-hot encoding increases the dimensionality of the data set.
    
    Ordinal encoding
    Similar to label encoding, but it allows you to explicitly define the mapping between categories and integer labels. This technique is best when there is a clear and predefined ordinal relationship.

```python
#Label Encoding: Label encoding is a technique used in machine learning and data analysis to convert categorical variables into numerical format.
#Pros: Simple and space-efficient
#Cons: Implies ordinal relationship,which may mislead models  {An ordinal relationship is a logical ranking or ordering of categories in a categorical variable. 
#The categories have a clear order, but the distances between them are not necessarily equal}
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
df.head()
```

```python
#One-Hot Encoding: One-hot encoding is a technique that converts categorical data into a format that machine learning algorithms can use to improve predictions
#Pros: Great for non-ordinal categories,maintain all category information
#Cons: Increases dimensionality,particularly with many unique values
df=pd.get_dummies(df,columns=['City'],drop_first=True)
df.head()
```

```python
df=createdata()
```

```python
#Ordinal encoding : Ordinal encoding is a technique to transform categorical features into a numerical format. In ordinal encoding, labels are translated to numbers based on their ordinal relationship to one another.
#pros : effective for ordinal data,captures order
#cons :not suitabble for ordinal data as it imples ranking
df['City']=df['City'].map({'New York':1,'San Francisco':2,'Los Angeles':3})
df.head()
```

```python
#Target encoding : Target encoding is the process of replacing a categorical value with the mean of the target variable. 
#Pros : useful for high cardinality features
#Cons : ca cause data leakage if target encoding is not done properly
mean_target=df.groupby('City')['Purchased'].mean()
df['City_encoded']=df['City'].map(mean_target)
df.head()
```

When you need to scale features
Use StandardScaler when data is normally distributed and you want to center features with mean 0 and standard deviation 1, which is ideal for linear models and PCA. 

MinMaxScaler is best for scaling data to a fixed range, often [0, 1], and works well for algorithms sensitive to feature ranges (like neural networks). 

RobustScaler is optimal for data with outliers, as it uses the median and interquartile range for scaling, making it more resistant to outliers than StandardScaler and MinMaxScaler.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

#Standardization
#Pros : works well for normally distributed data suitaable for many models
#Cons : Senstive to outliers
scaler=StandardScaler()
df[['Age','Salary']]=scaler.fit_transform(df[['Age','Salary']])
df.head()
```

```python
df=createdata()
df.head()
```

```python
#Normalization (range 0-1)
#Pros : keeps all data between 0 and 1;ideal for distance-based model
#Cons : can disort data distribution with more outliers
normalizer=MinMaxScaler()
df[['Age','Salary']]=normalizer.fit_transform(df[['Age','Salary']])
df.head()
```

```python
df=createdata()
df.head()
```

```python
#Robust Scaling (uses median and IQR, robust to outliers)
#Pros : more robust to outliers by using median and IQR
#Cons : sont work well with normal distribution
robust_scaler=RobustScaler()
df[['Age','Salary']]=robust_scaler.fit_transform(df[['Age','Salary']])
df.head()
```

When you have outliers

Z-Score: Calculates how many standard deviations each data point is from the mean. Typically, values with an absolute Z-score above a certain threshold (e.g., 3) are considered outliers. This method works well when data is normally distributed.

IQR (Interquartile Range): Calculates the range between the 25th percentile (Q1) and the 75th percentile (Q3) and defines outliers as data points below Q11.5IQRQ11.5IQR or above Q3+1.5IQRQ3+1.5IQR. IQR is effective for skewed distributions and is robust to outliers.

Median Replacement: Replaces outliers with the median value of the dataset. This method preserves the structure of the data and is useful when you want to keep the dataset size the same while reducing the impact of extreme values.

```python
# Outlier Detection and Treatment using IQR
#Pros: Simple and effective for mild outliers.
#Cons: May overly reduce variation if there are many extreme outliers.
Q1 = df['Salary'].quantile (0.25)
Q3 = df['Salary'].quantile(0.75)
IQR=Q3 - Q1
lower_bound = Q1 - 1.5* IQR
upper_bound = Q3 + 1.5 * IQR
df['Salary'] = np.where (df['Salary'] > upper_bound, upper_bound, np.where(df ['Salary'] < lower_bound, lower_bound, df ['Salary']))
```

```python
df=createdata()
df.head()
```

```python
#Z-score method
#Pros: Good for normally distributed data.
#Cons: Not suitable for non-normal data; may miss outliers in skewed distributions.
from scipy import stats
df['Salary_zscore'] = stats.zscore (df ['Salary'])
df['Salary'] = np.where (df['Salary_zscore'].abs() > 3, np.nan, df ['Salary']) # Replace outliers with NaN
```

```python
df=createdata()
df.head()
```

```python
# Median replacement for outliers
#Pros: Keeps distribution shape intact, useful when capping isn't feasible.
#Cons: May distort data if outliers represent real phenomena.
df['Salary_zscore'] = stats.zscore (df ['Salary'])
median_salary = df['Salary'].median()
df['Salary'] = np.where (df['Salary_zscore'].abs() > 3, median_salary, df['Salary'])
df.head(5)
```

Feature Engineering

Binning: Converts continuous features into categorical ones.
Purpose: Simplifies data, can reduce noise, and might help certain models like decision trees that perform better with categorical features.

Feature Combinations: Make new features by adding, subtracting, or dividing existing ones (e.g., BMI from weight/height22). This can reveal relationships and improve model accuracy.

Polynomial Features: Create new features by raising existing ones to powers (e.g., x2x2, x3x3) or combining them (e.g., x1x2x1x2) to capture non-linear relationships.

```python
df=createdata()
df.head()
```

```python
#Creating a new feature based on Salary (Binning Technique)
#Pros: Simplifies continuous features; useful in non-linear models.
#Cons: Can lose data granularity, potentially reducing model accuracy.
df['Income_Level'] = pd.cut(df['Salary'], bins=[0, 50000, 100000, 150000], labels=['Low', 'Medium', 'High'])
df.head(10)
```

```python
# Polynomial Features
#Pros: Captures complex relationships between variables.
#Cons: Increases dimensionality, risking overfitting.
from sklearn.preprocessing import PolynomialFeatures
poly= PolynomialFeatures(degree=2, include_bias=False)
poly_features= poly.fit_transform(df [['Age', 'Salary']])
df_poly = pd.DataFrame (poly_features, columns=poly.get_feature_names_out(['Age', 'Salary']))
df= pd.concat([df, df_poly], axis=1)
df.head()
```
